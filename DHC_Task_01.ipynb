## Task No 01:-

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
```

```python
print("Shape of the DataFrame:", df.shape)
print("Column Names:" , df.columns.tolist())
print("First few rows of the DataFrame:" )
print(df.head())
```

## Summary Statistics

```python
print("Statistical information about the data:")
print(df.info())
print("Descriptive statistics of the DataFrame:")
print(df.describe())
```

## Visualizations:-

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df)
plt.title('PetalLength vs PetalWidth')
plt.show()
```

```python
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=df)
plt.title('SepalLength vs PetalLength')
plt.show()
```

```python
sns.scatterplot(x='SepalWidthCm', y='PetalWidthCm', hue='Species', data=df)
plt.title('SepalWidth vs PetalWidth')
plt.show()
```

## HistPlot:-

```python
df.hist(figsize=(10, 6), edgecolor='black')
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, palette="Set2")
plt.title("Boxplots of Iris Features")
plt.show()
```

```python
sns.boxplot(x="Species", y="SepalLengthCm", data=df)
plt.title("Sepal Length by Species")
plt.show()
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=df)
plt.title("Petal Length by Species")
plt.show()
```
